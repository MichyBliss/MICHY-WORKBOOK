Fill the missing values in the year_resale_value with the average year_resale_value, then drop all other rows with missing values in the dataset


```python
# Calculate the mean of the 'year_resale_value' column
mean_year_resale_value = df['year_resale_value'].mean()
mean_year_resale_value
```

```python
df.columns
```

```python
df['__year_resale_value'].fillna(df['__year_resale_value'].mean(), inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Clean the name of the __year_resale_value column by removing the leading underscore
df.rename(columns={'__year_resale_value': 'year_resale_value'}, inplace=True)
```

```python
# Do descriptive statistics (mean, median, standard deviation) of the numerical columns. What do you notice? or i will use df.describe


# Compute mean
mean_values = df.mean(numeric_only=True)

# Compute median
median_values = df.median(numeric_only=True)

# Compute standard deviation
std_values = df.std(numeric_only=True)

# Combine into one DataFrame for clarity
summary_stats = pd.DataFrame({
    'Mean': mean_values,
    'Median': median_values,
    'Standard Deviation': std_values
})

print(summary_stats)
```

```python
# Check if the dataset has any duplicates. If there are, drop them
duplicates = df.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")
```

```python
# Find out the most common car brand in the dataset using `value_counts()`
df['Manufacturer'].value_counts()
```

```python
# Which manufacturer has the highest and lowest average sales volume?
manufacturer_sales = df.groupby('Manufacturer')['Sales_in_thousands'].mean()
manufacturer_sales

sorted_sales = manufacturer_sales.sort_values(ascending=False)
print(sorted_sales)
```

```python
# What is the distribution of car prices in the dataset?
plt.figure(figsize=(8, 6))
sns.histplot(df['Price_in_thousands'], bins=20, kde=True)
plt.title('Distribution of Car Prices')
plt.xlabel('Manufacturer')
plt.ylabel('Frequency')
```

```python
# Plot a correlation matrix to see if different numerical variables in the dataset correlate with each other
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

```python
# Plot a histogram of the car prices using Matplotlib
plt.figure(figsize=(8, 6))
plt.hist(df['Price_in_thousands'], bins=20, edgecolor='k')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
# Convert a column to datetime
df['Latest_Launch'] = pd.to_datetime(df['Latest_Launch'])
```

```python
# Plot the relationship between price and latest launch year.
plt.figure(figsize=(10, 6))
sns.lineplot(data = df, x='Latest_Launch', y='Price_in_thousands', marker='o')
plt.title('Average Car Price by Latest Launch Year')
plt.xlabel('Latest_Launch')
plt.ylabel('Price_in_thousands')
plt.grid(True)
plt.show()
```

```python
# Group cars by manufacturers and analyze their average price and fuel efficiency
manufacturer_stats = df.groupby('Manufacturer').agg({
    'Price_in_thousands': 'mean',
    'Fuel_efficiency': 'mean'
})
manufacturer_stats
```

```python
# Can you identify and visualize the most popular car brands?
popular_brands = df['Manufacturer'].value_counts().head(10)
popular_brands
```

```python
# Which car model has the highest resale value compared to its initial price?
#Calculate the highest resale value in percentage
df['Resale_Value_Percentage'] = (df['year_resale_value'] / df['Price_in_thousands']) * 100

# Find the row with the highest resale value percentage
top_car = df.loc[df['Resale_Value_Percentage'].idxmax()]

# Display Model, Manufacturer, and Resale Percentage
print("Car with Highest Resale Value:")
print(f"Manufacturer: {top_car['Manufacturer']}")
print(f"Model: {top_car['Model']}")
print(f"Resale Value Percentage: {top_car['Resale_Value_Percentage']:.2f}%")
```

```python
# Find the top 3 fuel-efficient cars with an engine size above 2.5 liters.
top_fuel_efficient_cars = df[df['Engine_size'] > 2.5].nlargest(3, 'Fuel_efficiency')
top_fuel_efficient_cars
```
